### Problema del XOR:

```python
import numpy as np
from mlp.mlp_main import MLP
from mlp_sgd.MLP import SgdMLP
import matplotlib.pyplot as plt
```

```python
inputs_bits = np.array(
    [
        [0, 0],
        [1, 0],
        [0, 1],
        [1, 1],
    ]
)

output_XOR = np.array([0, 1, 1, 0])
```

#### MLP with SGD:

```python
mlp_sgd = SgdMLP(
    layers=[(2, 3, "leaky_relu"), (3, 2, "leaky_relu"), (2, 1, "sigmoid")],
    learning_rate=0.5,
)
```

```python
error = mlp_sgd.train(inputs_bits, output_XOR, epochs=5000, error_tolerance=0.01)
```

```python
for x, y in zip(inputs_bits, output_XOR):
    y_pred = mlp_sgd.predict(x)
    binary = 1 if y_pred[0] >= 0.5 else 0
    print(f"Input: {x} | Esperado: {y} | Predicho: {binary}")
```

```python
# Plot error trace
plt.plot(error)
plt.xlabel("Epoch")
plt.ylabel("Mean Squared Error")
plt.title("Training Error")
plt.show()
```

### MLP with Backpropagation implementation

```python
# Instancia
mlp = MLP(
    layers=[(2, 3, "leaky_relu"), (3, 2, "leaky_relu"), (2, 1, "sigmoid")],
    learning_rate=0.5,
)
```

```python
error_mlp = mlp.train(inputs_bits, output_XOR, epochs=5000, error_tolerance=0.01, batch_size=4)
```

```python
plt.plot(error_mlp)
plt.title("Error a lo largo de las épocas")
plt.xlabel("Época")
plt.ylabel("Error (MSE)")
plt.show()
```
